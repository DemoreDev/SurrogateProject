```python
import pandas as pd
```

```python
# Lê 3 resultados 
dfs = [
pd.read_csv("../experiments_results/raw/raw_birds_results.csv"),
pd.read_csv("../experiments_results/raw/raw_medical_results.csv"),
pd.read_csv("../experiments_results/raw/raw_scene_results.csv")
]

# Ordena por maior R2 para F1 (Quão bem o modelo explica a coluna F1)
dfs = [df.sort_values(by="R2 para F1 (macro averaged by label)", ascending=False) for df in dfs]
for df in dfs:
    display(df.head(5))
```

#### Aparentemente os melhores modelos variam entre xgboost, lightgbm e catboost, alternando a abordagem entre multi_output e regressor_chain. Possíveis explicações:
- Gradient Boosting Decision Trees (GBDT) são significativamente melhores que modelos lineares ou Redes Neurais em dados tabulares;
- O Boosting cria fronteiras de decisão complexas que capturam melhor a relação não-linear;
- O espaço de busca destes modelos é muito rico, então o Optuna consegue encontrar ótimos pontos de regularização (λ,γ) 
- A diferença entre as abordagens "multi_output" e "regressor_chain" é apenas ruído estatístico. Neste caso, critérios como simplicidade, tempo de treinamento, tamanho do modelo e tempo de inferência devem ser levados em conta.

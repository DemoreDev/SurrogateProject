```python
import pandas as pd

def show_pipelines(dataset):
    """
    Lê o meta-df e exibe a configuração completa (algoritmos e hiperparâmetros)
    dos 5 pipelines com o maior F1-Score, separados por categoria (FP, MEKA, WEKA).
    """
    data_path = f"data/meta/meta_processed/meta_proc_{dataset}.csv"
    df = pd.read_csv(data_path)
    
    print(f"Extraindo o Top 5 pipelines do dataframe: {dataset.upper()}\n")

    # Ordena o dataframe pelo F1 Score em ordem decrescente e pega os 5 melhores
    top5 = df.sort_values(by='F1 (macro averaged by label)', ascending=False).head(5)

    # Filtra apenas as colunas de configuração
    config_cols = df.columns[2:-3]

    for i, (idx, row) in enumerate(top5.iterrows(), 1):
        f1 = row['F1 (macro averaged by label)']
        print(f"=== {i}º Lugar (F1: {f1:.4f}) ===")
        
        # Pega os algoritmos ativos (colunas sem '-' e com valor 1)
        active_algos = [c for c in config_cols if '-' not in c and row[c] == 1]
        
        # Dicionário para organizar a exibição
        pipeline_categorizado = {"FP": [], "MEKA": [], "WEKA": []}
        
        for algo in active_algos:
            # Extrai hiperparâmetros
            algo_params = [c for c in config_cols if c.startswith(f"{algo}-")]
            params_dict = {}
            for param in algo_params:
                val = row[param]
                if pd.notna(val) and val != -1:
                    param_name = param.split('-')[-1]
                    params_dict[param_name] = val
            
            # Categoriza o algoritmo
            if "mlfs" in algo or "sklearn" in algo.lower():
                pipeline_categorizado["FP"].append((algo, params_dict))
            elif "meka.classifiers" in algo:
                pipeline_categorizado["MEKA"].append((algo, params_dict))
            elif "weka.classifiers" in algo:
                pipeline_categorizado["WEKA"].append((algo, params_dict))
        
        # Imprime na ordem correta do fluxo (FP -> MEKA -> WEKA)
        for categoria in ["FP", "MEKA", "WEKA"]:
            algos_na_categoria = pipeline_categorizado[categoria]
            
            if not algos_na_categoria:
                print(f"  [{categoria}] Nenhum algoritmo utilizado")
            else:
                for algo, params in algos_na_categoria:
                    param_str = params if params else "(Default/Nenhum)"
                    print(f"  [{categoria}] {algo} | Params: {param_str}")
                
        print("-" * 70)
```
